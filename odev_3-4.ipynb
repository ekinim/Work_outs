```python
import numpy as np
import matplotlib.pyplot as plt
import math

from scipy.stats import norm
```

```python
ras1=np.random.normal(100, 15, 1000)
ras2=np.random.normal(100, 15, 10000)

ras1.sort()
ras2.sort()

plt.hist(ras1, density=True, bins=np.arange(40,160), color="r", alpha=0.3)
plt.hist(ras2, density=True, bins=np.arange(40, 160), color="b", alpha=0.3)
plt.show()
```

```python
p1=np.random.binomial(1, 0.5, 1000)
p2=np.random.binomial(1, 0.75, 1000)

a1=[]
a2=[]
sayi1=[]
sayi2=[]

for i in p1:
    sayi1.append(i)
    ort1=np.mean(sayi1)
    a1.append(ort1)
    
for j in p2:
    sayi2.append(j)
    ort2=np.mean(sayi2)
    a2.append(ort2)

plt.plot(a1, color="b")
plt.plot(a2, color="r")
plt.show()
```
