```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
```

```python
mnist = fetch_openml('Fashion-MNIST', version=1, cache=True)
```

```python
np.random.seed(286)
index = np.random.choice(70000,10000)

X = mnist.data[index]

y = mnist.target[index]
y=[int(i) for i in y]
```

```python
X_std=StandardScaler().fit_transform(X)
```

```python
tsne=TSNE(n_components=2, verbose=1, perplexity=30, n_iter=300).fit_transform(X_std)
```

```python
plt.figure(figsize=(15,8))

plt.scatter(tsne.T[0], tsne.T[1], c=y)
plt.show()
```

Some features are intertwined into eachother and it is impossible to distinguish them.

Some features are seperated from others and can easily be distinguished.

Results are better than PCA, but yet not good enough to use.

```python
tsne=TSNE(n_components=2, verbose=1, perplexity=50, n_iter=300).fit_transform(X_std)
```

```python
plt.figure(figsize=(15,8))

plt.scatter(tsne.T[0], tsne.T[1], c=y)
plt.show()
```

With 50 perplexity value, results are a little bit better than 30 perplexity value.
