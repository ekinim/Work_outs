```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
rest=pd.read_excel("C:/Users/Gunesim/Desktop/Ödevler/restaurants.xls")

a=["Düşük Fiyatlı", "Ortalama", "Pahalı"]

b=((np.sum(rest["price"]=="low")) & (np.sum(rest["franchise"]=="t")),
    (np.sum(rest["price"]=="medium")) & (np.sum(rest["franchise"]=="t")),
    (np.sum(rest["price"]=="high")) & (np.sum(rest["franchise"]=="t")))

c=((np.sum(rest["price"]=="low")) & (np.sum(rest["franchise"]=="f")),
    (np.sum(rest["price"]=="medium")) & (np.sum(rest["franchise"]=="f")),
    (np.sum(rest["price"]=="high")) & (np.sum(rest["franchise"]=="f")))

plt.subplot(1,1,1)
plt.bar(a, b, 0.3)

frc=plt.subplot(1,1,1)
nfrc=plt.bar(a, c, 0.3, bottom=b)

plt.ylabel("RESTORAN SAYISI")
plt.xlabel("FİYATLAR")
plt.title("MEKSİKA'DAKİ RESTORANLARIN FRANCHISE DURUMLARI")
plt.legend(("Franchise", "Franchise değil"))

plt.show()
```
