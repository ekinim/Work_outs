```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
veri=pd.DataFrame()
veri["Fiyat"]=np.random.normal(1000, 50, 100).astype(np.int)
veri["Satış Miktarı"]=np.random.normal(50, 5, 100).astype(np.int)
veri["Kâr Miktarı"]=np.random.normal(100, 10, 100).astype(np.int)
```

```python
#egilim hesaplama aracı ----------------------------------------------------
def egilim(liste):
    egil=[]
    ortalama=0
    medyan=0
    mod=[]
    mod_tablo=pd.DataFrame()
    liste=np.sort(liste)
    
    #ortalama hesabı ----------------------------------------------------
    ortalama=round(liste.sum()/len(liste), 2)

    #medyan hesabı ----------------------------------------------------
    if len(liste)%2!=0:
        medyan=round(liste[(len(liste)+1)/2], 2)
    else:
        medyan=round((liste[int(len(liste)/2)] + liste[int(len(liste)/2+1)])/2, 2)
    
    #mod hesabı ----------------------------------------------------
    mod_tablo[0]=np.unique(liste)
    mod_tablo[1]=0
    
    sayac=0
    for j in mod_tablo[0]:
        for k in liste:
            if j==k:
                mod_tablo[1][sayac]+=1
        sayac+=1
    
    mod_sayisi=np.max(mod_tablo[1])
    
    sayac=0
    for l in mod_tablo[1]:
        if l==mod_sayisi:
            mod.append(mod_tablo[0][sayac])
        sayac+=1

    #değerleri listeye yükleme ve geri döndürme ----------------------------------------------------
    egil=[ortalama, medyan, mod]
    return egil

#dagilim hesaplama aracı ----------------------------------------------------
def dagilim(liste):
    dagil=[]
    ort=np.mean(liste)
    varyans=0
    sapma=0
    hata=0
    
    #varyans hesabı ----------------------------------------------------
    for i in liste:
        varyans+=((i-ort)**2)/(len(liste)-1)
    varyans=round(varyans, 2)
    #standart sapma hesabı ----------------------------------------------------
    sapma=round(np.sqrt(varyans), 2)
    
    #standart hata hesabı ----------------------------------------------------
    hata=round(sapma/np.sqrt(len(liste)), 2)

    #değerleri listeye yükleme ve geri döndürme ----------------------------------------------------
    dagil=[varyans, sapma, hata]
    return dagil
```

```python
(satis, m2) = np.unique(veri['Satış Miktarı'], return_counts=True)
a2=np.argmax(m2)
mod_sat=satis[a2]

oto_sat1=[veri.mean()[1], veri.median()[1], mod_sat]
oto_sat2=[round(veri.var(ddof=1)[1], 2), round(veri.std()[1], 2), round(veri.sem()[1], 2)]
man_sat1=egilim(veri["Satış Miktarı"])
man_sat2=dagilim(veri["Satış Miktarı"])

bilgiler=[[man_sat1[0], man_sat1[1], man_sat1[2], man_sat2[0], man_sat2[1], man_sat2[2]],
          [oto_sat1[0], oto_sat1[1], oto_sat1[2], oto_sat2[0], oto_sat2[1], oto_sat2[2]]]
```

```python
bilgiler=pd.DataFrame(bilgiler)
bilgiler.columns=["Ortalama", "Medyan", "Mod", "Varyans", "S.Sapma", "S.Hata"]
bilgiler.index=["Manuel Hesaplanan", "Numpy İle Hesaplanan"]
bilgiler.index.name="Satış Miktarı Verisinin Değerleri"

bilgiler
```
